Q1

```python
import random

def f(x):
  return -x**2 + 6*x

x = random.randint(0, 6)
print(f"Initial x = {x}, f(x) = {f(x)}")

while True:
  current = f(x)
  neighbors = [n for n in [x - 1, x + 1] if 0 <= n <= 6]
  best = max(neighbors, key=f)

  if f(best) > current:
    x = best
    print(f"Moved to x = {x}, f(x) = {f(x)}")
  else:
    break

print(f"Final optimal x = {x}, f(x) = {f(x)}")
```

Q2

```python
goal = 20
k = 2
beam_width = 2

def h(n):
  return abs(goal - n)

def get_next(n):
  return [n + 2, n + 3, n * 2]

beam = [(1, [1])]
level = 0
found_path = None

while beam:
  print(f"Level {level}: {[state for state, _ in beam]}")

  if any(state == goal for state, _ in beam):
    for state, path in beam:
      if state == goal:
        found_path = path
        break
    break

  candidates = []
  for state, path in beam:
    for nxt in get_next(state):
      if nxt <= goal:
        candidates.append((nxt, path + [nxt]))

  if not candidates:
    break

  candidates.sort(key=lambda x: h(x[0]))
  beam = candidates[:beam_width]
  level += 1

print(f"Final path to reach {goal}: {found_path}")
```

```python
import random

POP_SIZE = 6
BITS = 5
GENERATIONS = 15
MUTATION_RATE = 0.01

def fitness(x):
  return x**2 + 2*x

def decode(chrom):
  return int("".join(map(str, chrom)), 2)

def random_chrom():
  return [random.randint(0, 1) for _ in range(BITS)]

def select(pop):
  fits = [fitness(decode(c)) for c in pop]
  total = sum(fits)
  pick = random.uniform(0, total)
  running = 0
  for c, f in zip(pop, fits):
    running += f
    if running >= pick:
      return c
  return pop[-1]

def crossover(p1, p2):
  pt = random.randint(1, BITS - 1)
  return p1[:pt] + p2[pt:], p2[:pt] + p1[pt:]

def mutate(chrom):
  return [bit ^ 1 if random.random() < MUTATION_RATE else bit for bit in chrom]

population = [random_chrom() for _ in range(POP_SIZE)]

for gen in range(GENERATIONS):
  new_pop = []
  while len(new_pop) < POP_SIZE:
    p1, p2 = select(population), select(population)
    c1, c2 = crossover(p1, p2)
    new_pop += [mutate(c1), mutate(c2)]
  population = new_pop[:POP_SIZE]

best = max(population, key=lambda c: fitness(decode(c)))
best_x = decode(best)
print(f"Best chromosome: {''.join(map(str, best))}")
print(f"Best x = {best_x}")
print(f"Best fitness f({best_x}) = {fitness(best_x)}")
```

```python
name = input("Enter student name: ")
marks = input("Enter marks: ")

grade = ''

if marks >= 85:
  grade = 'A'
elif marks >= 70:
  grade = 'B'
elif marks >= 50:
  grade = 'C'
else:
  grade = 'F'

print("Grade is " + grade)
```

```python
n = int(input("Enter n: "))

if n > 0:
  count = 0

  print("Even numbers: ")
  for i in range(n):
    if (i + 1) % 2 == 0:
      count += 1
      print(i + 1)

  print("Count: " + str(count))
```

Q5

```python
def calculate_average(marks_list):
  sum = 0
  for marks in marks_list:
    sum += marks

  return sum / len(marks_list)

n = int(input("Enter number of marks in list: "))

marks_list = []

for i in range(n):
  marks = int(input("Marks " + str(i) + ": "))
  marks_list.append(marks)

print("Avg: " + str(calculate_average(marks_list)))
```

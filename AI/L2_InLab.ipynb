```python
class System:
  def __init__(self):
    self.lights = {}

  def switch_light(self, room_name, onOrOff):
    self.lights[room_name] = onOrOff

  def get_status(self):
    for (k, v) in self.lights.items():
      print(f"Room '{k}': {"On" if v else "Off"}")

  def create_light(self, room_name):
    self.lights[room_name] = False

sys = System()

sys.create_light('Bedroom')
sys.create_light('Bathroom')
sys.create_light('Office')

sys.switch_light('Bedroom', True)

sys.get_status()
```

```python
class Staff:
  def __init__(self, name, staff_id, dept):
    self.name = name
    self.staff_id = staff_id
    self.dept = dept

  def display_info(self):
    print("ID:", self.staff_id)
    print("Name:", self.name)
    print("Dept:", self.dept)

class Teacher(Staff):
  def __init__(self, name, staff_id, dept, courses, salary):
    super().__init__(name, staff_id, dept)
    self.courses = courses
    self.salary = salary

  def teach(self):
    print(self.name, "is teaching")

  def display_info(self):
    super().display_info()
    print("Courses:", self.courses)
    print("Salary:", self.salary)


class Admin(Staff):
  def __init__(self, name, staff_id, dept, role, working_hours):
    super().__init__(name, staff_id, dept)
    self.role = role
    self.working_hours = working_hours

  def perform_task(self):
    print(self.name, "is working")

  def display_info(self):
    super().display_info()
    print("Role:", self.role)
    print("Working Hrs:", self.working_hours)


class Research(Staff):
  def __init__(self, name, staff_id, dept, research_topic, stipend):
    super().__init__(name, staff_id, dept)
    self.research_topic = research_topic
    self.stipend = stipend

  def work_on_research(self):
    print(self.name, "is researching")

  def display_info(self):
    super().display_info()
    print("Research Topic:", self.research_topic)
    print("Stipend:", self.stipend)

ad = Admin("Fadil", 635, "CS", "Database Admin", 35)
ad.display_info()
ad.perform_task()

print()

tch = Teacher("Ahmad", 1010, "CS", ["OOP"], 125000)
tch.display_info()
tch.teach()

print()

rch = Research("Zain", 818, "AI", "Big Data", 40000)
rch.display_info()
rch.work_on_research()
```

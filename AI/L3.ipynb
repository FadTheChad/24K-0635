Q1

```python
class TrafficLightAgent:
  def act(self, percept):
    if percept == "Heavy Traffic":
      return "Extend Green Time"
    elif percept == "Light Traffic":
      return "Normal Green"

agent = TrafficLightAgent()

for state in ["Heavy Traffic", "Light Traffic"]:
  action = agent.act(state)
  print("Percept:", state, "→ Action:", action)
```

Q2

```python
import random

class SmartLightAgent:
  def __init__(self):
    self.prev_presence = None
    self.prev_light = "OFF"

  def act(self, presence):
    action = "No action"
    if presence == "Yes" and self.prev_light == "OFF":
      action = "Turn lights ON"
      self.prev_light = "ON"
    elif presence == "No" and self.prev_light == "ON":
      action = "Turn lights OFF"
      self.prev_light = "OFF"
    self.prev_presence = presence
    return action

agent = SmartLightAgent()

for step in range(1, 9):
  presence = random.choice(["Yes", "No"])
  action = agent.act(presence)
  print("Step", step, "Students:", presence, "Action:", action)
  print("Model -> Previous Presence:", agent.prev_presence, "Light Status:", agent.prev_light)
```

```python
restaurants = {
  "A": {"Distance": 3, "Rating": 7},
  "B": {"Distance": 5, "Rating": 9}
}

utilities = {}

for name, data in restaurants.items():
  utility = data["Rating"] - data["Distance"]
  utilities[name] = utility
  print("Restaurant", name, "Utility =", utility)

selected = max(utilities, key=utilities.get)
print("Selected Restaurant:", selected)
```

Q5

```python
import random

actions = ["Play", "Rest"]
rewards = {"Play": 5, "Rest": 1}
q_table = {"Play": 0, "Rest": 0}
alpha = 0.5

for step in range(1, 11):
  action = random.choice(actions)
  reward = rewards[action]
  q_table[action] = q_table[action] + alpha * (reward - q_table[action])
  print("Step", step, "Action", action, "Reward", reward)

print("Q-table Updated")
print(q_table)
```

```python
grid = {
  "a": "safe", "b": "safe", "c": "fire",
  "d": "safe", "e": "fire", "f": "safe",
  "g": "safe", "h": "safe", "j": "fire"
}

path = ["a", "b", "c", "d", "e", "f", "g", "h", "j"]

def display():
  rooms = list(grid.keys())
  for i in range(0, 9, 3):
    row = rooms[i:i+3]
    for r in row:
      if grid[r] == "fire":
        print("F", end=" ")
      else:
        print(".", end=" ")
    print()
  print()

for room in path:
  print("Robot moving to room", room)
  if grid[room] == "fire":
    print("Fire detected in room", room, "- Extinguishing")
    grid[room] = "safe"
  else:
    print("No fire in room", room)
  display()

print("Final Environment Status")
display()
```
